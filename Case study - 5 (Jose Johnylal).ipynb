##  Do all the zones generate the same amount of sales?

To answer this question, we have to perform a one-way ANOVA test

> * Null hypothesis: All zones generate the same amount of sales
> * Alternate hypothesis: Atleast one zone has a difference in the amount of sales

```python
f, p = stats.f_oneway(sale['Zone - A'], sale['Zone - B'], sale['Zone - C'], sale['Zone - D'])
f, p
```

Since the p-value is less than 0.05 we can reject the null hypothesis and hence conclude that **All the zones do not generate the same amount of sales.**
